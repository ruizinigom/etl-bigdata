##Paso 1: Transformar 1 línea con Python

```python
#AUTOR: ÍÑIGO RUIZ MARTÍNEZ
#FECHA: 25/05/2022
#OBJETIVO: Conseguir de output: tipoCasa, servicio, nombre, opinion, vecindario, precio

#EXPLICACIÓN VARIABLES ELEGIDAS A TRANSFORMAR
#avaiability_365[-1]; - tipoCasa -'vacacional' si <= 100, 'segunda vivienda' si <=250, 'negocio si >250
#room_type[8]; - servicio - 1 si habitación privada, 0 si entire home, -1 otras
#name[1]; - nombre - sin manipulaciones
#reviewas_per_month[-3]; opinion - dividir entre 30 para sacar opiniones por día - float
#neighborhood[5]; -vecindario- 0:centro 1:salamanca 2:chamberí 3:embajadores 4:retiro -1:resto
#price[9]; precio - por día


lineaOrigen = '21853,Bright and airy room,83531,Abdel,Latina,Cármenes,40.40381,-3.7413,Private room,31,4,33,2018-07-15,0.42,2,364'
lineaTransformada = lineaOrigen.split(',')

tipoCasa = float(lineaTransformada[-1])
if (tipoCasa <= 100):
  tipoCasa = 'vacacional'
elif(tipoCasa > 100 and tipoCasa <= 250):
  tipoCasa = 'segunda vivienda'
elif(tipoCasa > 250):
  tipoCasa = 'negocio'
else:
  tipoCasa = -1

servicio = lineaTransformada[8]
if (servicio.lower() == 'private room'):
  servicio = 1
elif (servicio.lower()[:12] == 'entire home'):
  servicio = 0
else:
  servicio = -1

if not (lineaTransformada[1] == ''):
  nombre = lineaTransformada[1].strip()
else:
  nombre = -1

if not (lineaTransformada[-3] == ''):
  opinion = round(float(lineaTransformada[-3]) * 30,2)
else:
  opinion = -1

vecindario = lineaTransformada[4]
if (vecindario.lower() == 'centro'):
  vecindario = 0
elif(vecindario.lower() == 'salamanca'):
  vecindario = 1
elif(vecindario.lower() == 'chamberí'):
  vecindario = 2
elif(vecindario.lower() == 'latina'):
  vecindario = 3
elif(vecindario.lower() == 'retiro'):
  vecindario = 4
else:
  vecindario = -1

if not (lineaTransformada[9] == ''):
  precio = float(lineaTransformada[9])
else:
  precio = -1


lineaDestino = [tipoCasa, servicio, nombre, opinion, vecindario, precio]
print(lineaDestino)
```
